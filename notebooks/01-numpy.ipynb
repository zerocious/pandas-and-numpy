```python
## Задание 2.3
## Сгенерируйте матрицу 1000x3: суммы ~`N(500, 200), кол-во товаров[1, 10], возврат[0, 1]`.
## Отфильтруйте строки, где сумма > 800 И возврат == 0. Посчитайте их долю от общего числа.
## Создайте новый столбец сегмент:
    ## Premium если сумма > 1000 и кол-во > 5
    ## Standard если сумма > 600
    ## Low иначе
## Добавьте этот столбец к матрице как 4-ю колонку (результат будет объектным массивом, т.к. строки смешивают числа и текст).
import numpy as np

np.random.seed(0)
sums = np.random.normal(500, 200, size=1000)
qty = np.random.randint(1, 11, size=1000)
refund = np.random.choice([0, 1], size=1000, p=[0.9, 0.1])

data = np.column_stack((sums, qty, refund))

print(data[:5]) # просмотр данных

data_filtered = (data[:, 0] > 800) & (data[:, 2] == 0)
print(f"Доля: {data_filtered.mean() * 100}%")

conditions = [(data[:, 0] > 1000) & (data[:, 1] > 5), data[:, 0] > 600]
choices = ["Premium", "Standard"]
segments = np.select(conditions, choices, default = "Low")

data_obj = data.astype(object)
data_final = np.column_stack((data_obj, segments))
print(data_final[:5])
```

```python
## Задание 3.2
## Сгенерируйте массив из 1 000 000 случайных чисел N(0, 1).
## Напишите функцию с for, которая: если x > 0 → x**2, иначе abs(x). Замерьте время.
## Напишите векторный аналог через np.where. Замерьте время.
## Выведите отношение времени: time_loop / time_vec.
import numpy as np
import time as t

arr = np.random.randn(1_000_000)

def func(arr):
    arr_copy = np.copy(arr)
    for x in range(len(arr_copy)):
        if arr_copy[x] > 0:
            arr_copy[x] = arr_copy[x] ** 2
        else:
            arr_copy[x] = abs(arr_copy[x])
    return arr_copy

start = t.perf_counter()
arr_copy = func(arr)
end = t.perf_counter()
print(abs(start - end))

start = t.perf_counter()
arr_copy2 = np.where(arr > 0, arr ** 2, abs(arr))
end = t.perf_counter()
print(end - start)
```

```python
# 🎯 БОСС-ФАЙТ ДНЯ 1: «Векторная подготовка сырого датасета»

# 📋 ТЗ: 4 этапа обработки
# 1. Клиппинг выбросов
# Для каждого из 4 столбцов вычисли 95-й перцентиль (игнорируя NaN). Замени все значения > p95 на p95. NaN не трогать.
    
# 2. Импутация пропусков
# Замени все оставшиеся NaN на медиану соответствующего столбца.
    
# 3. Сегментация клиентов (приоритет сверху вниз)
# Создай 1D-массив segments длиной 5000 со строковыми метками:
#     "VIP": Income > 70000 И Spend > 3000
#     "Active": Spend > 2500 И Rating > 7
#     "AtRisk": Income < 30000 И Age > 45
#     "Standard": остальные

# 4. Min-Max нормализация:
#     Приведи все 4 числовых столбца к диапазону [0, 1] формулой (x-min)/(max-min). Вычисления делай по столбцам векторно.
    
# ⛔ Жёсткие ограничения:
#     0 циклов for/while/list comprehensions.
#     Только numpy.
#     Время выполнения < 0.3 сек

# 📦 Формат сдачи
# Выведи в консоль:
#     Shape: (5000, 4)
#     NaN count: 0
#     Min/Max per col: [0.0, 1.0] (approx)
#     Segments: {'VIP': X, 'Active': Y, 'AtRisk': Z, 'Standard': W}
#     Time: 0.XX sec

import time, numpy as np

start = time.perf_counter()
np.random.seed(42)
data = np.column_stack([
    np.random.normal(35, 12, 5000),       # Age
    np.random.normal(50_000, 15_000, 5000), # Income
    np.random.normal(2_000, 800, 5000),   # Spend
    np.random.uniform(1, 10, 5000)        # Rating
])
data[np.random.rand(5000, 4) < 0.05] = np.nan
data[np.random.rand(5000, 4) < 0.02] *= 10

# 1. Клиппинг выбросов
p_95 = np.nanpercentile(data, 95, axis=0)
data_clipped = np.where(data > p_95, p_95, data)

# 2. Импутация пропусков
data_imit = np.where(np.isnan(data_clipped), np.nanmedian(data[:], axis=0), data_clipped)

# 3. Сегментация клиентов (приоритет сверху вниз)
conds = [(data_imit[:, 1] > 70000) & (data_imit[:, 2] > 3000), (data_imit[:, 2] > 2500) & (data_imit[:, 3] > 7), (data_imit[:, 1] < 30000) & (data_imit[:, 0] > 45)]
choices = ["VIP", "Active", "AtRisk"]
segments = np.select(conds, choices, default="Standard")

# 4. Min-Max нормализация:
data_imit = (data_imit[:] - np.min(data_imit, axis=0)) / (np.max(data_imit, axis=0) - np.min(data_imit, axis=0))
end = time.perf_counter()

def prints():
    dict_for_segments = {'VIP': np.sum(np.where(segments == choices[0], 1, 0)), 'Active': np.sum(np.where(segments == choices[1], 1, 0)), 'AtRisk': np.sum(np.where(segments == choices[2], 1, 0)), 'Standard': np.sum(np.where(segments == "Standard", 1, 0))}
    print(f"Shape: {np.shape(data_imit)}", f"NaN count: {np.sum(np.isnan(data_imit))}", f"Min/Max per col: {np.min(data_imit, axis=0)}, {np.max(data_imit, axis=0)} (approx)", f"Segments: {dict_for_segments}", f"Time: {np.round(end - start, 2)} sec", sep = "\n")
    
prints()
```
